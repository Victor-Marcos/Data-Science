```python
import numpy as np
from sklearn import preprocessing, neighbors
import pandas as pd
```

```python
import pandas as pd
url = 'https://github.com/joanby/python-ml-course/raw/master/datasets/cancer/breast-cancer-wisconsin.data.txt'
df = pd.read_csv(url, header = None)
df.head()
```

```python
df.describe()
```

Si faltaron datos, se ponen datos muy grandes para evitar que afectaran a los resultados

```python
df.columns = ["name", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9", "class"]
```

```python
df.head()
```

```python
df = df.drop(["name"], axis=1)
df.replace("?", -99999, inplace=True)
```

```python
Y = df["class"]
X = df[["V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9"]]
```

Clasificador de los K vecinos:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
accuracy
```

Es muy importante que todas as columnas sean relevantes para "calcular distancias":

```python
df = pd.read_csv(url, header = None)
df.replace("?", -99999, inplace=True)
df.columns = ["name", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9", "class"]

Y = df["class"]
X = df[["name", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9"]]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

clf2 = neighbors.KNeighborsClassifier()
clf2.fit(X_train, Y_train)

accuracy = clf2.score(X_test, Y_test)
accuracy
```
